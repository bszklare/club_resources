# 1. Functions
A function declaration has 5 parts:
1. The keyword `def`
2. A name
3. 0 or more arguments in a comma-separated list (surrounded by parenthases)
4. A colon (`:`)
5. A body

We put them together like this:
```python
def <name>(<args>):
    <body>
```

````python
def add_one(x): #takes an argument
    '''This is a docstring. Helpful information about a function.'''
    return x + 1 #return statement = what the function will give back
````

````python
add_one(5)
````

````python
def add_many(a,b,c=2,d=5): #a function can take optional arguments as shown here. you do not need to specify them when you call the function, but you can
    return a + b + c + d
````

````python
add_many(1,2,3,4)
````

````python
add_many(1,2)
````

````python
add_many(1,2,d=3)
````

### Practice Problem 1.
Write a function that takes in a stock price and an annual return and returns its value after 10 years of receiving that rate of return.

### <font color=red>Answer Key:</font>

````python
#key formula: final price = (initial price) * (1 + rate)^10
def ten_yr_price(stock_price, ann_return):
    one_yr_return = 1 + ann_return
    ten_yr_return = one_yr_return ** 10
    final_price = stock_price * ten_yr_return
    return final_price
````

````python
def ten_yr_price(stock_price, ann_return):
    return stock_price * ((1 + ann_return)**10)
````

````python
ten_yr_price(50,.07) #rule of 70: takes about 70/(r*100) periods to double. in this cause it is roughly 10.
````

### Practice Problem 2.

Write a function that simulates buying something if the price is less than 20, and selling if it's at least 20. (Hint) Have the function accept a price, and total initial cash reserve. 


We'll do this by: 
1. decreasing our cash reserve if the price is less than 20
2. increasing it otherwise
3. testing the function on a 25 dollar stock starting with 500 dollars.

### <font color=red>Answer Key:</font>

````python
def buy_or_sell(price, bankroll):
    if price < 20:
        bankroll -= price # buy
        action = 'buy'
    else:
        bankroll += price # sell
        action = 'sell'
    return bankroll,action
````

````python
buy_or_sell(25, 500)
````

# 3. Loops
### 2 Key Types of Loops: For loops and While loops

#### While Loops

````python
x = 0
while(x < 10): #while loop, continues until a condition is met
    print(x)
    x += 1
````

#### For Loops

````python
for i in range(10): #for loop. continues for a set number of iterations
    print(i)  #range function with the argument 10 informs the forloop to iterate over the integers 0 through 9
````

````python
zero_to_ten = [i for i in range(11)] #can put for loops inside lists
print(zero_to_ten)
````

````python
zero_to_fifteen = [] 
for number in range(16): #can print the contents of a list one-by-one
    zero_to_fifteen.append(number)
print(zero_to_fifteen)
````

````python
#looping through a basic dictionary's keys
basic_dict = {'a':1, 'b':2, 'c':3}
for k in basic_dict:
    print(basic_dict[k])
````

1. Do the following: 
    1. Print out the max price of this list, along with the ticker symbol for that stock.
    2. Print out the min prince of this list, along with the ticker symbol for that stock.
(hint) create a dictionary
2. Now suppose you have 350 dollars, and can either buy or sell each stock. You want to sell a stock if the price is at least 20, and buy it otherwise. Do the following:
    1. Find the remaining balance after considering each stock.
    2. Return a portfolio that provides which stocks have been bought and which were sold in the market.
(hint) use the function from practice problem 2.
     


### <font color=red>Answer Key:</font>

````python
market = {'AAPL':12.32, 'GOOG':31.88, 'AMZN':17.51, 'MSFT':45.04, 'NFLX':22.60}
max_price = market['AAPL']
min_price = market['AAPL']
max_price_ticker = 'AAPL'
min_price_ticker = 'AAPL'
for ticker in market:
    if market[ticker] > max_price:
        max_price = market[ticker]
        max_price_ticker = ticker
    elif market[ticker] < min_price:
        min_price = market[ticker]
        min_price_ticker = ticker

print('the highest priced stock is: ', max_price_ticker, max_price)
print('the lowest priced stock is: ', min_price_ticker, min_price)
````

````python
#Here is the function we made in practice problem 2.
def buy_or_sell(price, bankroll):
    if price < 20:
        bankroll -= price # buy
        action = 'buy'
    else:
        bankroll += price # sell
        action = 'sell'
    return bankroll, action
````

````python
bankroll = 350.0
portfolio = {}
for ticker in market:
    price = market[ticker]
    bankroll_act = buy_or_sell(price, bankroll)
    bankroll = bankroll_act[0]
    action = bankroll_act[1]
    portfolio[ticker] = action
print(portfolio,' with remaining cash =', bankroll)    
````

## <font color=green>Easy Example: Fizzbuzz</font>

Write code that checks the numbers 0 to 50 and:
- If the number is a multiple of 3, prints "fizz"
- If the number is a multiple of 5, prints "buzz"
- If the number is a multiple of 3 and 5, prints "fizzbuzz"
- Otherwise prints the number



````python
for i in range(50):
    if i % 3 == 0 and i % 5 == 0:
        print('fizzbuzz')
    elif i % 3 == 0:
        print('fizz')
    elif i % 5 == 0:
        print('buzz')
    else:
        print(i)
````

## <font color=orange>Intermediate Example</font>
- Say you have an array for which the *ith* element is the price of a given stock on day i.

 - If you were only permitted to complete at most one transaction (i.e., buy one and sell one share of the stock), design an algorithm to find the maximum profit.

- Note that you cannot sell a stock before you buy one.

#### EX1
- Input: [7,1,5,3,6,4]
- Output: 5
- Explanation: Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5.
             Not 7-1 = 6, as selling price needs to be larger than buying price.

#### EX 2
- Input: [7,6,4,3,1]
- Output: 0
- Explanation: In this case, no transaction is done, i.e. max profit = 0.
