# A More Advanced Example

Using some of the concepts from the introduction to python we put together a more complicated example. We simulate a stock for 1000 days by generating random returns. If you stick around, doing something like this will be super easy for you.

```python
# Import statistical packages
import numpy as np # handle arrays 
import matplotlib.pyplot as plt # plotting libraries
from scipy.stats import norm # statistics
%matplotlib inline

# Configure parameters
t_intervals = 1000
iterations = 5
stdev = 0.02
drift = 0.0005
initial_price = 1

# Format parameters as arrays (where necessary)
drift_arr = drift * np.ones((t_intervals, iterations))
stdev_arr = stdev * np.ones((t_intervals, iterations))
prices = initial_price * np.ones((t_intervals, iterations))

# Randomly simulate returns
daily_returns = np.exp(drift_arr + stdev_arr*norm.ppf(np.random.rand(t_intervals, iterations)))

# Compute daily stock prices
for time in range(1, t_intervals):
    prices[time] = prices[time-1] * daily_returns[time]

# Plot stock chart
plt.figure(figsize=(14,7))
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Stock')
plt.plot(prices);
```
